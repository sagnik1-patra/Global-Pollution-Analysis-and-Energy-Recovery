```python
import os
import json
import warnings
import joblib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import (
    r2_score,
    mean_squared_error,
    mean_absolute_error,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)

warnings.filterwarnings("ignore")

# ============================================================
# 1. Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery\Global_Pollution_Analysis.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================================
# 2. Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)

# Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

print("\nCleaned Columns:\n", df.columns)

# Save raw preview
df.head().to_csv(os.path.join(OUTPUT_DIR, "dataset_preview.csv"), index=False)

# ============================================================
# 3. Basic Dataset Info
# ============================================================

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

missing_values = df.isnull().sum()
missing_values.to_csv(os.path.join(OUTPUT_DIR, "missing_values.csv"))

# ============================================================
# 4. Handle Missing Values
# ============================================================

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")

df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])

if len(categorical_cols) > 0:
    df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])

print("\nMissing Values After Imputation:")
print(df.isnull().sum())

# Save cleaned dataset
df.to_csv(os.path.join(OUTPUT_DIR, "cleaned_global_pollution_data.csv"), index=False)

# ============================================================
# 5. Descriptive Statistics
# ============================================================

desc_stats = df.describe()
desc_stats.to_csv(os.path.join(OUTPUT_DIR, "descriptive_statistics.csv"))

print("\nDescriptive Statistics:")
print(desc_stats)

# ============================================================
# 6. Feature Engineering
# ============================================================

# Energy consumption per capita
if "Energy_Consumption_in_TWh" in df.columns and "Population_in_millions" in df.columns:
    df["Energy_Consumption_per_Capita"] = (
        df["Energy_Consumption_in_TWh"] / df["Population_in_millions"]
    )

# Pollution severity category
if "Air_Pollution_Index" in df.columns:
    df["Pollution_Severity"] = pd.cut(
        df["Air_Pollution_Index"],
        bins=[-np.inf, 50, 100, np.inf],
        labels=["Low", "Medium", "High"]
    )

# Total pollution score
pollution_cols = []

for col in [
    "Air_Pollution_Index",
    "Water_Pollution_Index",
    "Soil_Pollution_Index",
    "CO2_Emissions_in_MT",
    "Industrial_Waste_in_tons"
]:
    if col in df.columns:
        pollution_cols.append(col)

if len(pollution_cols) > 0:
    scaler_pollution = StandardScaler()
    pollution_scaled = scaler_pollution.fit_transform(df[pollution_cols])
    df["Total_Pollution_Score"] = pollution_scaled.mean(axis=1)

# Yearly trend features
if "Year" in df.columns:
    yearly_trends = df.groupby("Year").mean(numeric_only=True)
    yearly_trends.to_csv(os.path.join(OUTPUT_DIR, "yearly_trends.csv"))

df.to_csv(os.path.join(OUTPUT_DIR, "feature_engineered_data.csv"), index=False)

# ============================================================
# 7. Encoding Categorical Columns
# ============================================================

df_encoded = df.copy()

label_encoders = {}

for col in df_encoded.select_dtypes(include=["object", "category"]).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le

joblib.dump(label_encoders, os.path.join(OUTPUT_DIR, "label_encoders.pkl"))

df_encoded.to_csv(os.path.join(OUTPUT_DIR, "encoded_data.csv"), index=False)

# ============================================================
# 8. Correlation Analysis
# ============================================================

correlation_matrix = df_encoded.corr(numeric_only=True)
correlation_matrix.to_csv(os.path.join(OUTPUT_DIR, "correlation_matrix.csv"))

plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, aspect="auto")
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "correlation_heatmap.png"), dpi=300)
plt.close()

# ============================================================
# 9. Visualizations
# ============================================================

# Bar chart: average air pollution by country
if "Country" in df.columns and "Air_Pollution_Index" in df.columns:
    avg_air_pollution = (
        df.groupby("Country")["Air_Pollution_Index"]
        .mean()
        .sort_values(ascending=False)
    )

    avg_air_pollution.to_csv(os.path.join(OUTPUT_DIR, "average_air_pollution_by_country.csv"))

    plt.figure(figsize=(14, 6))
    avg_air_pollution.plot(kind="bar")
    plt.title("Average Air Pollution Index by Country")
    plt.xlabel("Country")
    plt.ylabel("Average Air Pollution Index")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "average_air_pollution_by_country.png"), dpi=300)
    plt.close()

# Line plot: energy recovered over year
if "Year" in df.columns and "Energy_Recovered_in_GWh" in df.columns:
    yearly_energy = df.groupby("Year")["Energy_Recovered_in_GWh"].mean()

    yearly_energy.to_csv(os.path.join(OUTPUT_DIR, "yearly_energy_recovered.csv"))

    plt.figure(figsize=(10, 5))
    yearly_energy.plot(kind="line", marker="o")
    plt.title("Average Energy Recovered Over Years")
    plt.xlabel("Year")
    plt.ylabel("Energy Recovered in GWh")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "yearly_energy_recovered.png"), dpi=300)
    plt.close()

# Box plot: CO2 emissions by severity
if "Pollution_Severity" in df.columns and "CO2_Emissions_in_MT" in df.columns:
    plt.figure(figsize=(8, 5))
    df.boxplot(column="CO2_Emissions_in_MT", by="Pollution_Severity")
    plt.title("CO2 Emissions by Pollution Severity")
    plt.suptitle("")
    plt.xlabel("Pollution Severity")
    plt.ylabel("CO2 Emissions in MT")
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "co2_emissions_by_pollution_severity.png"), dpi=300)
    plt.close()

# Scatter plot: pollution vs energy recovery
if "Air_Pollution_Index" in df.columns and "Energy_Recovered_in_GWh" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["Air_Pollution_Index"], df["Energy_Recovered_in_GWh"])
    plt.title("Air Pollution Index vs Energy Recovered")
    plt.xlabel("Air Pollution Index")
    plt.ylabel("Energy Recovered in GWh")
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "air_pollution_vs_energy_recovered.png"), dpi=300)
    plt.close()

# ============================================================
# 10. Linear Regression Model
# Objective: Predict Energy Recovered in GWh
# ============================================================

target_regression = "Energy_Recovered_in_GWh"

regression_features = [
    "Air_Pollution_Index",
    "Water_Pollution_Index",
    "Soil_Pollution_Index",
    "CO2_Emissions_in_MT",
    "Industrial_Waste_in_tons",
    "Energy_Consumption_in_TWh",
    "Population_in_millions",
    "GDP_in_Billion_USD",
    "Renewable_Energy_in_%"
]

regression_features = [
    col for col in regression_features
    if col in df_encoded.columns and col != target_regression
]

X_reg = df_encoded[regression_features]
y_reg = df_encoded[target_regression]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

linear_model = LinearRegression()
linear_model.fit(X_train_reg_scaled, y_train_reg)

y_pred_reg = linear_model.predict(X_test_reg_scaled)

r2 = r2_score(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)

regression_metrics = {
    "Model": "Linear Regression",
    "R2_Score": r2,
    "MSE": mse,
    "MAE": mae,
    "RMSE": rmse
}

pd.DataFrame([regression_metrics]).to_csv(
    os.path.join(OUTPUT_DIR, "linear_regression_metrics.csv"),
    index=False
)

print("\nLinear Regression Metrics:")
print(regression_metrics)

# Save predictions
regression_predictions = pd.DataFrame({
    "Actual_Energy_Recovered_in_GWh": y_test_reg.values,
    "Predicted_Energy_Recovered_in_GWh": y_pred_reg
})

regression_predictions.to_csv(
    os.path.join(OUTPUT_DIR, "linear_regression_predictions.csv"),
    index=False
)

# Actual vs predicted graph
plt.figure(figsize=(8, 5))
plt.scatter(y_test_reg, y_pred_reg)
plt.xlabel("Actual Energy Recovered in GWh")
plt.ylabel("Predicted Energy Recovered in GWh")
plt.title("Linear Regression: Actual vs Predicted")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "linear_regression_actual_vs_predicted.png"), dpi=300)
plt.close()

# Save model and scaler
joblib.dump(linear_model, os.path.join(OUTPUT_DIR, "linear_regression_model.pkl"))
joblib.dump(scaler_reg, os.path.join(OUTPUT_DIR, "linear_regression_scaler.pkl"))

# ============================================================
# 11. Logistic Regression Model
# Objective: Classify Pollution Severity
# ============================================================

target_classification = "Pollution_Severity"

classification_features = [
    "Air_Pollution_Index",
    "Water_Pollution_Index",
    "Soil_Pollution_Index",
    "CO2_Emissions_in_MT",
    "Industrial_Waste_in_tons",
    "Energy_Consumption_in_TWh",
    "Population_in_millions",
    "GDP_in_Billion_USD",
    "Renewable_Energy_in_%"
]

classification_features = [
    col for col in classification_features
    if col in df_encoded.columns
]

X_cls = df_encoded[classification_features]
y_cls = df_encoded[target_classification]

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls
)

scaler_cls = StandardScaler()
X_train_cls_scaled = scaler_cls.fit_transform(X_train_cls)
X_test_cls_scaled = scaler_cls.transform(X_test_cls)

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_cls_scaled, y_train_cls)

y_pred_cls = logistic_model.predict(X_test_cls_scaled)

accuracy = accuracy_score(y_test_cls, y_pred_cls)
precision = precision_score(y_test_cls, y_pred_cls, average="weighted")
recall = recall_score(y_test_cls, y_pred_cls, average="weighted")
f1 = f1_score(y_test_cls, y_pred_cls, average="weighted")

classification_metrics = {
    "Model": "Logistic Regression",
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1_Score": f1
}

pd.DataFrame([classification_metrics]).to_csv(
    os.path.join(OUTPUT_DIR, "logistic_regression_metrics.csv"),
    index=False
)

print("\nLogistic Regression Metrics:")
print(classification_metrics)

# Classification report
class_report = classification_report(y_test_cls, y_pred_cls, output_dict=True)

with open(os.path.join(OUTPUT_DIR, "classification_report.json"), "w") as f:
    json.dump(class_report, f, indent=4)

classification_report_df = pd.DataFrame(class_report).transpose()
classification_report_df.to_csv(
    os.path.join(OUTPUT_DIR, "classification_report.csv")
)

# Confusion matrix
cm = confusion_matrix(y_test_cls, y_pred_cls)
cm_df = pd.DataFrame(cm)
cm_df.to_csv(os.path.join(OUTPUT_DIR, "confusion_matrix.csv"), index=False)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Logistic Regression Confusion Matrix")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "logistic_regression_confusion_matrix.png"), dpi=300)
plt.close()

# Save classification predictions
classification_predictions = pd.DataFrame({
    "Actual_Pollution_Severity": y_test_cls.values,
    "Predicted_Pollution_Severity": y_pred_cls
})

classification_predictions.to_csv(
    os.path.join(OUTPUT_DIR, "logistic_regression_predictions.csv"),
    index=False
)

# Save model and scaler
joblib.dump(logistic_model, os.path.join(OUTPUT_DIR, "logistic_regression_model.pkl"))
joblib.dump(scaler_cls, os.path.join(OUTPUT_DIR, "logistic_regression_scaler.pkl"))

# ============================================================
# 12. Countries Needing Improvement
# ============================================================

if "Country" in df.columns:
    country_summary = df.groupby("Country").agg({
        "Air_Pollution_Index": "mean",
        "Water_Pollution_Index": "mean",
        "Soil_Pollution_Index": "mean",
        "CO2_Emissions_in_MT": "mean",
        "Industrial_Waste_in_tons": "mean",
        "Energy_Recovered_in_GWh": "mean"
    }).reset_index()

    country_summary["Improvement_Priority_Score"] = (
        country_summary["Air_Pollution_Index"]
        + country_summary["Water_Pollution_Index"]
        + country_summary["Soil_Pollution_Index"]
        + country_summary["CO2_Emissions_in_MT"]
        + country_summary["Industrial_Waste_in_tons"]
    ) / (country_summary["Energy_Recovered_in_GWh"] + 1)

    country_summary = country_summary.sort_values(
        by="Improvement_Priority_Score",
        ascending=False
    )

    country_summary.to_csv(
        os.path.join(OUTPUT_DIR, "countries_needing_improvement.csv"),
        index=False
    )

# ============================================================
# 13. Model Comparison
# ============================================================

model_comparison = pd.DataFrame([
    {
        "Model": "Linear Regression",
        "Task": "Energy Recovery Prediction",
        "Main_Metric": "R2 Score",
        "Score": r2,
        "Other_Metrics": f"MSE={mse}, MAE={mae}, RMSE={rmse}"
    },
    {
        "Model": "Logistic Regression",
        "Task": "Pollution Severity Classification",
        "Main_Metric": "Accuracy",
        "Score": accuracy,
        "Other_Metrics": f"Precision={precision}, Recall={recall}, F1={f1}"
    }
])

model_comparison.to_csv(
    os.path.join(OUTPUT_DIR, "model_comparison.csv"),
    index=False
)

plt.figure(figsize=(8, 5))
plt.bar(model_comparison["Model"], model_comparison["Score"])
plt.title("Model Comparison")
plt.ylabel("Score")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "model_comparison.png"), dpi=300)
plt.close()

# ============================================================
# 14. Final Report
# ============================================================

report = f"""
# Global Pollution Analysis and Energy Recovery Report

## Objective

The objective of this project is to analyze global pollution data and develop
machine learning models to predict energy recovery from pollution levels.
The project also classifies pollution severity and provides actionable insights
for pollution reduction and energy recovery improvement.

## Dataset

Dataset used:

Global_Pollution_Analysis.csv

Rows: {df.shape[0]}
Columns: {df.shape[1]}

## Data Preprocessing

The dataset was loaded and column names were cleaned.
Missing numerical values were handled using median imputation.
Missing categorical values were handled using most frequent value imputation.
Categorical columns were encoded using Label Encoding.
Pollution-related numerical features were scaled where required.

## Feature Engineering

The following features were created:

1. Energy Consumption per Capita
2. Pollution Severity Category
3. Total Pollution Score
4. Yearly Trend Summary

## Exploratory Data Analysis

EDA included:

1. Descriptive statistics
2. Missing value analysis
3. Correlation heatmap
4. Average air pollution by country
5. Energy recovery trends over years
6. CO2 emission distribution by pollution severity
7. Air pollution vs energy recovery scatter plot

## Linear Regression Model

Objective:

Predict Energy Recovered in GWh using pollution indicators and other features.

Metrics:

R2 Score: {r2}
MSE: {mse}
MAE: {mae}
RMSE: {rmse}

## Logistic Regression Model

Objective:

Classify pollution severity into Low, Medium, and High categories.

Metrics:

Accuracy: {accuracy}
Precision: {precision}
Recall: {recall}
F1 Score: {f1}

## Model Comparison

Linear Regression was used for continuous energy recovery prediction.
Logistic Regression was used for pollution severity classification.

Both models solve different objectives, so their scores are not directly
comparable. Linear Regression performance should be judged using R2, MSE,
MAE, and RMSE. Logistic Regression performance should be judged using
Accuracy, Precision, Recall, F1 Score, and Confusion Matrix.

## Actionable Insights

1. Countries with high pollution and low energy recovery should be prioritized.
2. Industrial waste can be converted into useful energy through waste-to-energy systems.
3. High CO2-emitting countries should increase renewable energy adoption.
4. Countries with poor air pollution scores should improve emission control policies.
5. Energy recovery can be improved by investing in recycling, biomass energy,
   biogas generation, and industrial waste heat recovery.
6. Pollution severity classification can help policymakers identify countries
   requiring urgent intervention.

## Recommendations

1. Adopt waste-to-energy conversion plants.
2. Improve industrial waste segregation and recycling.
3. Increase renewable energy usage.
4. Introduce stricter carbon emission policies.
5. Promote cleaner production technologies.
6. Monitor pollution indices yearly.
7. Use predictive models for future environmental planning.

## Generated Deliverables

1. Cleaned dataset
2. Encoded dataset
3. Feature engineered dataset
4. Descriptive statistics CSV
5. Missing value report
6. Correlation matrix
7. EDA graphs
8. Linear Regression model and metrics
9. Logistic Regression model and metrics
10. Confusion matrix
11. Model comparison file
12. Countries needing improvement file
13. Trained model files
"""

with open(os.path.join(OUTPUT_DIR, "final_report.md"), "w", encoding="utf-8") as f:
    f.write(report)

# ============================================================
# 15. Save Final Summary JSON
# ============================================================

summary = {
    "dataset_rows": int(df.shape[0]),
    "dataset_columns": int(df.shape[1]),
    "linear_regression": regression_metrics,
    "logistic_regression": classification_metrics,
    "output_directory": OUTPUT_DIR
}

with open(os.path.join(OUTPUT_DIR, "final_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\nProject Completed Successfully!")
print("All results saved at:")
print(OUTPUT_DIR)
```
